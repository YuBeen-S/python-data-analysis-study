### CSV 파일을 데이터프레임으로 읽기 (read_csv)

```python
import pandas as pd
df = pd.read_csv('../data/남산도서관 장서 대출목록 (2021년 4월).csv', encoding='cp949', low_memory=False)
# low_memory=False 옵션 미사용 시, DtypeWarning 발생 가능. (자동으로 파악한 데이터 타입이 달라지면 경고 발생)
# low_memory=False 옵션은 파일을 나누어 읽지 않고 한 번에 읽는 것.

df.head()
```

### 데이터프레임을 CSV 파일로 저장하기 (to_csv)

```python
df.to_csv('ns_202104.csv')
```

```python
with open('ns_202104.csv', encoding='utf-8') as f:
    for i in range(3):
        print(f.readline(), end='')
```
