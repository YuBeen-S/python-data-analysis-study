## JSON (JavaScript Object Notation)

### 파이썬 객체를 JSON 문자열로 변환 (json.dumps)

```python
d = {"name" : "혼자 공부하는 데이터 분석"}
print(d['name'])
```

```python
pip install json
```

```python
import json

d_str = json.dumps(d, ensure_ascii=False) # ensure_ascii=False 옵션은 한글이 깨지지 않도록 함
# d_str은 JSON 문자열
print(d_str)
```

```python
print(type(d_str))
```

### JSON 문자열을 파이썬 객체로 변환 (json.loads)

```python
d2 = json.loads(d_str) # JSON 문자열을 파이썬 객체로 변환
print(d2['name'])
```

```python
print(type(d2))
```

### json 패키지 이용한 변환 활용

```python
d3 = json.loads('{"name":"혼자 공부하는 데이터 분석", "author":"박해선", "year":2022}')
print(d3['name'])
print(d3['author'])
print(d3['year'])
```

```python
d3 = json.loads('{"name":"혼자 공부하는 데이터 분석", "author":["박해선", "홍길동"], "year":2022}')
print(d3['author'][1])
```

```python
d4_str = """
[
    {"name":"혼자 공부하는 데이터 분석", "author":"박해선", "year":2022},
    {"name":"혼자 공부하는 머신러닝+딥러닝", "author":"박해선", "year":2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]['name'])
```

### JSON 문자열을 DataFrame으로 변환 (read_json)

```python
from io import StringIO
import pandas as pd
pd.read_json(StringIO(d4_str)) # d4_str 문자열을 DataFrame으로 변환
```

```python
pd.DataFrame(d4) # d4 리스트(파이썬 객체)를 DataFrame으로 변환
```

## XML (eXtensible Markup Language)

### XML 문자열을 파이썬 객체로 변환 (fromstring)

```python
x_str = """
<book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
</book>
"""
```

```python
pip install xml.etree.ElementTree
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
print(type(book))

print(book.tag)
```

### 자식 엘리먼트 확인 (findtext)

```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

### 여러 개의 자식 엘리먼트 확인 (findall, for문)

```python
x2_str = """
<books>
    <book>
        <name>혼자 공부하는 데이터 분석</name>
        <author>박해선</author>
        <year>2022</year>
    </book>
    <book>
        <name>혼자 공부하는 머신러닝+딥러닝</name>
        <author>박해선</author>
        <year>2020</year>
    </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
for book in books.findall('book'):
    name = book.findtext('name')
    author = book.findtext('author')
    year = book.findtext('year')
    print(name)
    print(author)
    print(year)
    print()
```
